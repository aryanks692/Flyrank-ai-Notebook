## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
# This cell is for CODE (numbers, a query, a check).
## 2. Target or proxy

##**Target:** `is_declining_label`

##The target is to predict whether a content page is declining in performance.

##The target label is created from the observed `trend_direction` column in the starter dataset. Pages with a `trend_direction` of **"down"** are labeled as 1 (declining), while all other pages are labeled as 0 (not declining).

##This target represents an observed outcome in the dataset and is used only as the prediction label, not as an input feature.

# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All
# Create the target label

df["is_declining_label"] = (
    df["trend_direction"] == "down"
).astype(int)

# Display the target
df[["trend_direction", "is_declining_label"]].head(10)
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
# This cell is for CODE (numbers, a query, a check).
## 3. Success metric

##**Success Metric:** Precision@50

##Precision@50 measures how many of the top 50 pages predicted as declining are actually declining.

##This metric is appropriate because content editors can only refresh a limited number of pages. A high Precision@50 means the model identifies the most important pages for editors to review first.

##Secondary evaluation metrics such as ROC-AUC and Average Precision can also be used to measure the model's overall classification performance.
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Show the proportion of each class
df["is_declining_label"].value_counts(normalize=True)
```
