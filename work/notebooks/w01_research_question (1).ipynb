## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
## Research Question

##Question:
##Can historical search performance metrics help identify webpages that are likely to decline in search performance?

##Unit of analysis:
##One webpage.

##Decision:
##Which webpages should an SEO team review first?

##Action:
##The SEO team can inspect high-risk pages, refresh content, improve titles, update internal links, or optimize keywords.

##Cost of a wrong recommendation:
##A false positive means time is spent reviewing a page that may not decline. A false negative means a declining page is missed, which could result in reduced search traffic. Therefore, the model should support prioritization rather than replace human judgment.
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

print("Total pages:", len(df))

print("\nTrend direction counts:")
print(df["trend_direction"].value_counts())

print("\nAverage CTR:", round(df["ctr"].mean(), 2))
print("Average Position:", round(df["avg_position"].mean(), 2))
print("Average Content Age (days):", round(df["content_age_days"].mean(), 1))
## Quick Look at the Data

##The starter dataset contains **30,000 webpages**, providing a large sample for analysis.

##The `trend_direction` column shows that **16,262 pages are trending down**, **5,962 are stable**, **4,388 are trending up**, **2,236 are new**, and **1,152 are flat**. This suggests that there are many pages that may require review or content updates.

##The average click-through rate (CTR) is **0.51**, the average search position is **16.34**, and the average content age is **256.2 days**. These metrics indicate that the dataset contains historical search performance and content characteristics that can be used to support decisions about which pages should be reviewed first.
```
