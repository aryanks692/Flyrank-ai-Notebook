#### The pattern behind the score

Two things jump out that change the diagnosis:

**1. Search volume and impressions don't match.** Most of the top-priority pages target keywords with tiny search volume (10–140/mo) but are pulling in 5,000–60,000 impressions over 90 days. That means they're ranking for a wide spread of long-tail queries well beyond their intended keyword — and CTR on that traffic is dismal (0.02%–0.5%, against an expected 2–4% at position 10–20). That combination — decent position, high impressions, near-zero clicks — usually points to one of:
- The page ranks for queries it doesn't actually answer well (content-to-query mismatch), or

- Title tag and meta description aren't compelling enough to earn the click that the ranking has already won.

**2. A third of the highest-priority pages were updated in just the last 25–30 days and are *still* collapsing** (trend of -60% to -98%). This is the category most likely to hide a technical fault: wrong canonical, mismatched indexed metadata, or a different URL actually serving the SERP snippet than the one being tracked. 114 of these overlap directly with the High Business Impact group — same pages, different lens.

**Stale Content — 6,572 pages**
Untouched 91+ days with real impression volume. Root cause is straightforward content decay: outdated facts, thinner coverage than newer competitor content, no schema/format updates to compete for SERP features. About 2,000 of these overlap with the Declining Performance group, confirming staleness is a real contributor to some (not all) of the decline.

#### Step 6: Ranked by Impact vs. Effort

1. **Technical SEO / High Business Impact fixes (highest priority)** — largest dollar impact, lowest effort (metadata and technical checks, not full rewrites). Includes content_8451fc6f034d (~$192K/mo) and content_44e481c8f55b (~$12K/mo), both at position 1–3 with CTR under 1% of benchmark.

#### Step 7: Executive Summary

Three findings matter most: first, a small set of already well-ranked pages are leaking an outsized share of revenue — over $560K/month combined — due to CTR failures that look technical rather than content-related, and should be the team's first stop. Second, nearly 4,500 pages are in decline despite good positioning, and a meaningful subset of those were refreshed recently without improvement, meaning the fix is diagnostic (cannibalization, indexing) rather than another content edit.
